## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib as mpl
%matplotlib inline
import matplotlib.pyplot as plt

# print how many rows are there in the table named 'master'
QUERY = """
SELECT *
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df1 = pd.read_sql_query(QUERY,conn)
plt.figure(figsize=(20,6))
plt.hist(df1['ConvertedComp'], color = 'orange')
plt.title('Histogram for CovertedComp', size=25)
plt.ylabel('Frequency')
plt.xlabel('ConvertedComp')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
import seaborn as sns

plt.figure(figsize=(20,6))
sns.set(style="white")
sns.boxplot(df1['Age'], palette="Set1")
plt.title('Boxplot for Age', size=25)
plt.show()
df1['Age'].dtypes
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
plt.figure(figsize=(20,6))
plt.scatter(df1['Age'], df1['WorkWeekHrs'], color='red' )
plt.title('Scatter plot of Age and WorkWeekHrs', size=25)
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()
```

```python
# Correlation of Age and other variables
df1.corr()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

plt.figure(figsize=(20,6))
sns.scatterplot(data=df1, x="CodeRevHrs", y="WorkWeekHrs", size="Age", hue="Age", palette="hls", legend=True, sizes=(20, 2000))

# show the graph
plt.title('CodeRevHrs and WorkWeekHrs  By Age', size=25) 
plt.xlabel('CodeRevHrs') 
plt.ylabel('WorkWeekHrs')
plt.show()

```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

# print how many rows are there in the table named 'master'
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df2 = pd.read_sql_query(QUERY,conn)
df3 = df2['DatabaseDesireNextYear'].value_counts().rename_axis('DatabaseDesireNextYear').reset_index(name='Number')
#Only taking the top 5
dfp = df3.head(5)
#Calculating the percentage
dfp['percent']=((dfp['Number']/dfp['Number'].sum())*100)  

```

```python
#Rounding off the calculated percentage column
dfp['percent'] = dfp['percent'].round(decimals = 2)
#Adding the percentage sign
dfp['percent'] = dfp['percent'].astype(str) + '%'

```

```python
plt.figure(figsize=(20,6))
dfp['Number'].plot(kind='pie', labels=dfp['percent'])

plt.legend(labels=dfp['DatabaseDesireNextYear'], loc='upper right')
plt.title('Percentages of Top 5 Databases Respondents who wish To Learn',  size=25) 
plt.axis('equal')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
WHERE Age BETWEEN 30 AND 35
"""
df4 = pd.read_sql_query(QUERY,conn)

df5 = df4.groupby('Age').median()

df5.plot(kind ='bar', figsize=(20,6),stacked=True)
plt.legend( loc='lower right')
plt.title('Sacked chart of median WorkWeekHrs and CodeRevHrs for the age group 30 to 35.', size=25)
plt.ylabel('WorkWeekHrs')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
df_l=df1[['ConvertedComp', 'Age']]
df_l=df_l[df_l['Age']>=45]
df_l = df_l[df_l['Age']<=60]
df_l=df_l.groupby('Age').median()

df_l.plot(kind ='line', figsize=(20,6))
plt.title('Line graph for median ConvertedComp for all ages from 45 to 60.', size=25)
plt.ylabel('ConvertedComp')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
sns.set(style="white")
dd =df1['MainBranch'].value_counts().to_frame()

dd.plot(kind ='barh', color='grey', figsize=(20,6))
plt.title('horizontal bar chart using column MainBranch.', size=25)
plt.xlabel('Numbers')
plt.show()
```
